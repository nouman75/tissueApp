```python
import numpy as np
import pandas as pd
from sklearn.datasets import make_classification
from sklearn.feature_selection import SelectKBest, f_classif, SelectFromModel, RFE
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression

# Generate sample data
X, y = make_classification(n_samples=1000, n_features=20, n_informative=10, n_redundant=5, random_state=42)

# Univariate Feature Selection (SelectKBest with ANOVA F-test)
k = 10  # number of top features to select
selector = SelectKBest(score_func=f_classif, k=k)
X_new_univariate = selector.fit_transform(X, y)

# Tree-based Feature Selection (SelectFromModel with Random Forest)
clf = RandomForestClassifier(n_estimators=100, random_state=42)
selector = SelectFromModel(clf)
X_new_tree_based = selector.fit_transform(X, y)

# Greedy Feature Selection (RFE with Logistic Regression)
estimator = LogisticRegression(max_iter=1000, random_state=42)
selector = RFE(estimator, n_features_to_select=k)
X_new_greedy = selector.fit_transform(X, y)

# Print the selected features for each method
print("Selected features using Univariate Feature Selection:")
print(selector.get_support(indices=True))

print("Selected features using Tree-based Feature Selection:")
print(selector.get_support(indices=True))

print("Selected features using Greedy Feature Selection:")
print(selector.get_support(indices=True))
```

```python
import pandas as pd
import numpy as np

# Create a sample dataset with outliers
data = {
    'EmployeeID': range(1, 101),
    'DailyRate': np.random.randint(100, 500, size=100)  # Generating random DailyRate values
}
data['DailyRate'][0] = 2000  # Introducing an outlier

# Creating DataFrame
df = pd.DataFrame(data)

# Identifying and handling outliers
columns_with_outliers = ['DailyRate']

for c in columns_with_outliers:
    transform = 'clipped_' + c
    lower_limit = df[c].quantile(0.10)
    upper_limit = df[c].quantile(0.90)
    df[transform] = df[c].clip(lower_limit, upper_limit)

# Displaying the original and transformed DataFrame
print("Original DataFrame:")
print(df.head())

# Showing clipped values
print("\nTransformed DataFrame with clipped values:")
print(df[['DailyRate', 'clipped_DailyRate']].head())
```
